```python
def mystemp():
    import time
    
    today = time.localtime()
    
    InternDate = str(today.tm_year) + '-' + str(today.tm_mon) + '-' + str(today.tm_mday)
    InternTime = str(today.tm_hour) + ':' + str(today.tm_min) + ':' + str(today.tm_sec)
    InternDateFormat = InternDate + 'T' + InternTime
    return InternDateFormat

print(mystemp())
help(mystemp)
```

```python
#interpolation and extrapolation

import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

A = np.zeros([22, 2]) # 22 rows, 2 columns
A[:, 0] = np.arange(0, 43, 2) # all rows, first column, 2 - step size
A[0:11, 1] = [2, 6, 9, 12, 14, 16, 17.5, 18.5, 20, 20.5, 21.5]
A[11:22, 1] = [22, 22.5, 22.7, 23.5, 23.5, 23.7, 24, 24, 24.2, 24.2, 24.5]
plt.plot(A[:, 0], A[:, 1], 'co--', label='Measures')
plt.xlabel('Time [s]')
plt.ylabel('Votage [V]')

p2 = interpolate.lagrange(A[[0, 10, 21], 0], A[[0, 10, 21], 1]) # coordinates from rows 0, 10, 21
xnew = np.arange(-2, 50, 2)
ynew = p2(xnew)
error = ((p2(A[:,0]) - A[:, 1] ) ** 2).sum()
print('P2 => quadratic error: %.4e; calculated %.4e.' % (error, error/22))
plt.plot(xnew,ynew, 'g-', label='Polynomial order 2')

p5 = interpolate.lagrange(A[0:22:4, 0], A[0:22:4, 1])
xnew = np.arange(-2, 50, 2)
ynew = p5(xnew)
error = ((p5(A[:,0]) - A[:, 1] ) ** 2).sum()
print('P5 => quadratic error: %.4e; calculated %.4e.' % (error, error/22))
plt.plot(xnew,ynew, 'r-', label='Polynomial order 5')

plt.legend(loc='lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import time

plt.figure(1)
image = mpimg.imread('cat.png')
plt.imshow(image, cmap='gray')
print(image.shape)

newImage = np.copy(image)
t = time.time()

i = image.shape[0]
j = image.shape[1]

for x in range(0, i):
    for y in range(0, j):
        newImage[x, y, 0] = max(1 - (x/i*2 - 1)**2 - (y/j*2 - 1)**2, 0)
elapsed = time.time() - t
print("used time:", elapsed, "\n")
plt.figure(2)
plt.imshow(newImage)
```
